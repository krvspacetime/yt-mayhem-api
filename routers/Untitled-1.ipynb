```python
from yt_dlp import YoutubeDL
import pprint
```

```python
ydl_opts = {
        "skip_download": True,
        "quiet": False,
        "verbose": True,
    }
```

```python
def _get_format_info(format_id: str):
    with YoutubeDL({"quiet": True}) as ydl:
        info = ydl.extract_info(f"https://www.youtube.com/watch?v=WyiZYwdqAjQ", download=False)
        return next((f for f in info.get("formats", []) if f["format_id"] == format_id), None)

def _find_audio_format():
    with YoutubeDL({"quiet": True}) as ydl:
        info = ydl.extract_info(f"https://www.youtube.com/watch?v=WyiZYwdqAjQ", download=False)
        return next((f for f in info.get("formats", []) if f.get("acodec") != "none" and f.get("vcodec") == "none"), None)
```

```python
def get_info():
    with YoutubeDL({"quiet": True}) as ydl:
        info = ydl.extract_info(f"https://www.youtube.com/watch?v=WyiZYwdqAjQ", download=False)
        return info
```
